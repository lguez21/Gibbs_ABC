```python
import numpy as np
from scipy.stats import invgamma
```

```python
# Paramètres — valeurs exactes Section 10.2 du supplément
n          = 3
T_len      = 100
alpha_true = np.array([1.0, 2.0, 3.0])
s_true     = np.array([1.0, 1.0])
RNG_SEED   = 42   # même graine → mêmes données dans les 3 notebooks
```

```python
# Simulation MA(2) : x(t) = y_t + mu1*y_{t-1} + mu2*y_{t-2}
def simulate_ma2(mu, T, sigma2, rng):
    y = rng.normal(0.0, np.sqrt(sigma2), size=T + 2)
    return y[2:] + mu[0] * y[1:-1] + mu[1] * y[:-2]
```

```python
# Génération des données x* — ne pas changer l'ordre des appels
rng         = np.random.default_rng(RNG_SEED)
betas_true  = rng.dirichlet(alpha_true, size=n)
mu_true     = np.column_stack([
    betas_true[:, 0] - betas_true[:, 1],
    2 * (betas_true[:, 0] + betas_true[:, 1]) - 1
])
sigma2_true = invgamma.rvs(s_true[0], scale=s_true[1],
                            size=n, random_state=rng)
X_obs       = [simulate_ma2(mu_true[j], T_len, sigma2_true[j], rng)
               for j in range(n)]
```

```python
 
# Vérification — doit être identique dans les 3 notebooks
print("Vérification :")
for j in range(n):
    print(f"  Série {j+1}: mu=({mu_true[j,0]:+.4f}, {mu_true[j,1]:+.4f})"
          f"  sigma2={sigma2_true[j]:.4f}  x_obs[0]={X_obs[j][0]:+.4f}")
```
